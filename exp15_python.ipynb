```python
# PART-A
# dataset
import pandas as pd
import numpy as np

data = {
'Product':['Laptop','Mobile','Tablet','Headphones','Smartwatch','Camera'],
'Price':[55000,20000,30000,2000,8000,45000],
'Units_Sold':[25,60,35,80,50,15],
'Discount':[10,15,12,20,18,8]
}

df = pd.DataFrame(data)
print(df)
```

```python
# min-max normalization (prize)

df['Price_MinMax'] = (df['Price'] - df['Price'].min()) / (df['Price'].max() - df['Price'].min())
print(df[['Product','Price','Price_MinMax']])

# (units_sold)
df['Units_Sold_MinMax'] = (df['Units_Sold'] - df['Units_Sold'].min()) / (df['Units_Sold'].max() - df['Units_Sold'].min())
print(df[['Product','Units_Sold','Units_Sold_MinMax']])

# (discount)
df['Discount_MinMax'] = (df['Discount'] - df['Discount'].min()) / (df['Discount'].max() - df['Discount'].min())
print(df[['Product','Discount','Discount_MinMax']])
```

```python
# normalize multiple columns  (calculating MinMax for all the input data in one go)

cols = ['Price','Units_Sold','Discount']
df_norm = (df[cols] - df[cols].min()) / (df[cols].max() - df[cols].min())
print(df_norm)
```

```python
# Z-score normalization (units_sold)

df['Units_ZScore'] = (df['Units_Sold'] - df['Units_Sold'].mean()) / df['Units_Sold'].std()
print(df[['Product','Units_Sold','Units_ZScore']])

# (discount)
df['Discount_ZScore'] = (df['Discount'] - df['Discount'].mean()) / df['Discount'].std()
print(df[['Product','Discount','Discount_ZScore']])

# (price)
df['Price_ZScore'] = (df['Price'] - df['Price'].mean()) / df['Price'].std()
print(df[['Product','Price','Price_ZScore']])
```

```python
# decimal scaling (price) (pickup hghest value)

df['Price_Decimal'] = df['Price'] / 100000
print(df[['Product','Price','Price_Decimal']])

# (units_sold)
df['Units_Sold_Decimal'] = df['Units_Sold'] / 100
print(df[['Product','Units_Sold','Units_Sold_Decimal']])

# (discount)
df['Discount_Decimal'] = df['Discount'] / 100
print(df[['Product','Discount','Discount_Decimal']])
```

```python
# PART-B
# dataset
import pandas as pd
import numpy as np
data = {
'Order_ID':[101,102,103,104,105,106],
'Customer_Gender':['Male','Female','Female','Male','Female','Male'],
'Payment_Method':['UPI','Credit Card','Debit Card','COD','UPI','Debit Card'],
'Product_Category':['Electronics','Clothing','Home','Electronics','Beauty','Electronics'],
'City':['Pune','Mumbai','Delhi','Bangalore','Pune','Hyderabad'],
'Order_Value':[4500,2300,3200,5200,1800,4100]
}
df1 = pd.DataFrame(data)
print(df1)
```

```python
# method 1: Label Encoding

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df1['Gender_Label'] = le.fit_transform(df1['Customer_Gender'])
display(df1)
df1['City_Label'] = le.fit_transform(df1['City'])
display(df1)
df1['Product_Category_Label'] = le.fit_transform(df1['Product_Category'])
display(df1)
df1['Payment_Method_Label'] = le.fit_transform(df1['Payment_Method'])
display(df1)
```

```python
# method 2: One-Hot Encoding (can apply in multiple columns)

df_encoded = pd.get_dummies(df1, columns=['Customer_Gender','City','Product_Category','Payment_Method'])
display(df_encoded)
```

```python
# method 3: Dummy Encoding (drops first column)

df_dummy = pd.get_dummies(df1, columns=['Customer_Gender','City','Product_Category','Payment_Method'], drop_first=True)
display(df_dummy)
```
